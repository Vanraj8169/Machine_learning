```python
# iterator is a function in python that helps to iterate efficiently and memory management

string = "hello"
iterator = iter(string)
first_char = next(iterator)
second_char = next(iterator)
print(first_char)
print(second_char)
```

```python
# Generator are a simpler way to create iterators. They use the yield keyword to produce a series of values lazily, which means they generate values on the fly and do not store them in memory.

def square(n):
    for i in range(3):
        yield i**2
```

```python
square(3)
```

```python
for i in square(3):
    print(i)
```

```python
# Application of Generator
def read_large_file(file_path):
    with open(file_path,'r') as file:
        for line in file:
            yield line 
```

```python
file_path = 'large.txt'
for line in read_large_file(file_path):
    print(line.strip())
```
