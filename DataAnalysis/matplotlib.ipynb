```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6]
y = [1,4,9,16,25,36]
plt.plot(x,y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Basic Line Plot')
```

```python
plt.plot(x,y,color='red',linestyle='--',marker='o',linewidth=2)
plt.grid(True)
```

```python
 x = [1,2,3,4,5,6]
y = [1,4,9,16,25,36]
z = [2,3,2,5,6,2]

plt.subplot(2,2,1)
plt.plot(x,y,color='green')
plt.title('Plot 1')

plt.subplot(2,2,2)
plt.plot(y,z,color='blue')
plt.title('Plot 2')

plt.subplot(2,2,3)
plt.plot(y,x,color='pink')
plt.title('Plot 3')

plt.subplot(2,2,4)
plt.plot(z,y,color='red')
plt.title('Plot 4')
```

```python
# Histogram
data = [1,2,2,3,4,5,6,7,7,3,4,4]
plt.hist(data,bins=5,color='orange',edgecolor='black')
```

```python
# Create a scatter plot
x = [ 1,2,3,4,5]
y = [2,3,5,6,7]

plt.scatter(x,y,color='red')
```

```python
## pie chart

labels = ['A','B','C','D']
sizes = [30,20,20,10]
colors=['gold','yellowgreen','lightcoral','lightskyblue']
explode=(0.2,0,0,0) # Move out the 1st slice

plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct='%1.1f%%',shadow=True)
```
